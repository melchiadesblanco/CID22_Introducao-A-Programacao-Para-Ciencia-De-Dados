## Leitura e análise inicial dos dados

Inicie um DataFrame a partir do arquivo `2017-02-01_156_-_Base_de_Dados_sample.csv`. Exiba algumas linhas e informações sobre os tipos identificados automaticamente pelo Pandas.

**Dica:** Caso o arquivo não seja separado por vírgulas, o Pandas não conseguirá reconhecer os campos adequadamente. Você precisará fornecer o parâmetro `sep=';'` para indicar o separador correto (neste caso, o `;`).

**Dica:** Aplicativos modernos tendem a armazenar arquivos no formato UTF-8, mas é comum encontrar arquivos codificados em outros formatos. Caso você tenha problemas para ler o arquivo, utilize o parâmetro `encoding='latin-1'` para selecionar o encoding correto (*latin-1* neste exemplo).

```python
# Resposta:

import pandas as pd

df_bds = pd.read_csv("../data/2017-02-01_156_-_Base_de_Dados_sample.csv", sep=';', encoding='latin-1',  header='infer')
df_bds = df_bds.set_index('SOLICITACAO')
```

Visualize algumas linhas de dados:

```python
# Resposta:
df_bds.sample(n=10)
```

Use o método `info` para exibir as colunas e os tipos identificados.

```python
# Resposta:
df_bds.info()
```

Estamos interessados apenas nos seguintes campos: 'DATA', 'HORARIO', 'SUBDIVISAO', 'BAIRRO_ASS', 'SEXO', 'DATA_NASC', 'RESPOSTA_FINAL'

Crie (ou substitua) um DataFrame contendo apenas os campos acima.

```python
# Resposta:

df_bds = df_bds[[ 'DATA', 'HORARIO', 'SUBDIVISAO', 'BAIRRO_ASS', 'SEXO', 'DATA_NASC', 'RESPOSTA_FINAL']]

df_bds
```

## Localizando e tratando valores inválidos

Exiba todas as linhas com valores em branco (*NaN*).

```python
# Resposta:
#Não é possivel exibir todas as linhas devido à limitação do Jupyter
df_bds[df_bds.isna().any(axis=1)].sample(n=10)
```

```python
#Quantidade de registros que possuem pelo menos um valor com Nan
i_qtd_NaN = df_bds[df_bds.isna().any(axis=1)].size
i_qtd_tot = df_bds.size
i_qtd_NaN
```

Exclua todas as linhas que contenham algum valor nulo (*NaN*). Verifique se todas as linhas com valores em branco foram excluídas corretamente.

```python
# Resposta:
df_bds.dropna(axis='index', how='any', inplace=True)
i_qtd_upd = df_bds.size

if df_bds[df_bds.isna().any(axis=1)].size == 0:
    print('Nenhuma linha com valor em branco (NaN) !\nDas {} linhas existentes no dataset, {} foram excluídas. Atualmente há {}'.format(i_qtd_tot, i_qtd_NaN, i_qtd_upd))
else:
    print('Aparentemente, ainda há {} linhas com valores NaN'.format(df_bds[df_bds.isna().any(axis=1)].size))    
```

Faça com que todos os nomes de bairros fiquem em letras minúsculas.

```python
# Resposta:
df_bds['BAIRRO_ASS'] = df_bds['BAIRRO_ASS'].str.lower()


df_bds['BAIRRO_ASS'].unique()
```

## Conversão de tipos, criação de coluna e escrita em CSV

Exiba os tipos das colunas do DataFrame.

```python
# Resposta:
df_bds.info()
```

- Converta os campos de data para o formato DateTime. 
- Crie uma nova coluna chamada HORA, contendo apenas o componete hora do horário da reclamação (dica: use os comandos str.split e str.get)
- Converta a coluna HORA para Inteiro

```python
# Resposta:
#Converta os campos de data para o formato DateTime.
df_bds['DATA'] = df_bds['DATA'].astype('datetime64[ns]')
df_bds['DATA_NASC'] = df_bds['DATA_NASC'].astype('datetime64[ns]')
```

```python
#Crie uma nova coluna chamada HORA, contendo apenas o componete hora do horário da reclamação 
#(dica: use os comandos str.split e str.get)

df_bds['HORA'] = df_bds['HORARIO'].str.split(pat=':').str.get(0)
```

```python
#Converta a coluna HORA para Inteiro
df_bds['HORA'] = df_bds['HORA'].astype('int64')
```

Crie uma coluna chamada `IDADE` contendo a diferença entre o ano da reclamação e o ano de nascimento da pessoa.

**Dica:** Para extrair o ano de uma coluna do tipo data, use a propriedade `dt.year`. Por exemplo: `df['DATA'].dt.year`.

```python
# Resposta:
df_bds['IDADE'] = df_bds['DATA'].dt.year - df_bds['DATA_NASC'].dt.year

df_bds.sample(n=10)
```

Verifique se as colunas foram corretamente convertidas para os tipos corretos.

```python
# Resposta:
df_bds.info()
```

Identifique outliers na coluna `IDADE`. Remova do DataFrame todas as reclamações de pessoas com idades muito baixas.

```python
# Resposta:
#Identifique outliers na coluna IDADE
import matplotlib.pyplot as plt
import numpy as np

r = plt.boxplot(df_bds['IDADE'])
r
```

```python
outliers = r["fliers"][0].get_data()[1]
plt.plot(np.ones(len(outliers)), outliers, "+")
```

```python
outliers

### Notes: Verificando os outliers sugeridos pelo matplotlib, a parte inferior parece ser condizente com outliers
### no entanto a parte superior tenho dúvidas pois creio que pessoas de 90 anos ou mais podem ter feito reclamações.
### Por isso estarei removendo as idades 90+ da lista de outliers
```

```python
outliers = outliers[outliers < 90]
outliers
```

```python
#Remova do DataFrame todas as reclamações de pessoas com idades muito baixas.

df_bds = df_bds[df_bds['IDADE'].isin(outliers) == False]

r = plt.boxplot(df_bds['IDADE'])
r
```

Salve o DataFrame em um arquivo CSV chamado `../data/2017-02-01_156_-_Base_de_Dados_sample-limpo.csv`. Use o parâmetro `index=False` para não incluir o índice no aquivo gerado.

```python
# Resposta:
out_file = '../data/2017-02-01_156_-_Base_de_Dados_sample-limpo.csv'

df_bds.to_csv(out_file, index=False,  sep=';', encoding='UTF-8')
```
